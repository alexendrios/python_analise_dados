```python
#importando colunas específicas Hospital - Município - complexibilidade -carácter de atendimento - sub grupo de procedimento
#procedimento

df = pd.read_csv('sih-janeiro-2017-cirurgias-eletiva-e-emergencia.csv', sep=';', encoding='cp1252',
                usecols=[3, 6, 7, 8, 12, 14])

#trocar o nome das colunas
df.columns = ['Hospital', 'Município', 'Complexibilidade', 'Carater Adendimento',
            'Sub Grupo Procedimento', 'Procedimento']

df.head(3)
```

```python
df.describe()
```

```python
#listar os Hospitais presente nos dados

df['Hospital'].unique()
```

```python
#Quantidade de cirugias realizadas em cada Hospital
df['Hospital'].value_counts()
```

```python
#Quantas cirugias realizadas por subgrupo

df['Sub Grupo Procedimento'].value_counts()
```

```python
#plotar o Gráfico
df['Sub Grupo Procedimento'].value_counts().plot.bar()
```

```python
#dividir o dataframe original pelo procedimento de parto cesariano

df_parto_cesariano = df[df['Procedimento'] == 'PARTO CESARIANO']

#primeiras linhas

df_parto_cesariano.head()
```

```python
#ultimas linhas

df_parto_cesariano.tail(3)
```

```python
#Quantidade de Parto cesariano por Urgêncua

df_parto_cesariano['Carater Adendimento'].value_counts()
```

```python
#Quantidade de Parto cesariano por Hospital

df_parto_cesariano['Hospital'].value_counts()
```

```python
#Quantidade de Parto cesariano por Município

df_parto_cesariano['Município'].value_counts()
```

```python
#Quantidade de Parto cesariano por Complexidade

df_parto_cesariano['Complexibilidade'].value_counts()
```

```python
#Gráfico parto cesariana por município
df_parto_cesariano['Município'].value_counts().plot.barh()
```

```python
#iverter a ordem do gráfico e colocar um título no gráfico

df_parto_cesariano['Município'].value_counts(ascending=True).plot.barh(title='QUANTIDADES DE PARTOS CESARIANOS EM BRASÍLIA - DF')
```
